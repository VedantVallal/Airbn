```python
import pandas as pd
import numpy as np

print("Imported")
```

```python
data = {
    "Marks" : [30,32,404,45,50,60,70,90,300]
}

df =  pd.DataFrame(data)

Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)

```

```python
IQR = Q3-Q1

lower_bound = Q1-1.5*IQR
Upper_bound = Q3+1.5*IQR

Outlier = df[(df["Marks"] <= lower_bound) | (df["Marks"] >= Upper_bound)]
Pure_Data = df[(df["Marks"] >= lower_bound) & (df["Marks"] <= Upper_bound)]

print("Outlier : ")
print(Outlier)
print("Pure Data : ")
print(Pure_Data)
```
